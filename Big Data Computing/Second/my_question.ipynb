```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

# Configurazione dei dati
num_points_per_class = 4

# Definizione manuale dei punti per le due classi
class_0 = np.array([[-3, -3], [-3, -1], [-1, -3], [-1, -1]])

point_A = np.array([[-2, -2]])
point_B = np.array([[-6, 2]])

# Creazione delle etichette
X = class_0
y = [5, 5, 5, 5]

# Creazione e allenamento del modello KNN
k = 3
knn = KNeighborsClassifier(n_neighbors=k)
knn.fit(X, y)

# Creazione di una griglia per visualizzare i confini di decisione
x_min, x_max = X[:, 0].min() - 5, X[:, 0].max() + 5
y_min, y_max = X[:, 1].min() - 5, X[:, 1].max() + 5

# Plot dei dati e dei confini di decisione
plt.figure(figsize=(8, 6))
plt.scatter(class_0[:, 0], class_0[:, 1], color='blue', label='Classe 0', alpha=0.6)
plt.scatter(point_A[:, 0], point_A[:, 1], color='green', label='Punto A', marker='x', s=100)
plt.scatter(point_B[:, 0], point_B[:, 1], color='red', label='Punto B', marker='x', s=100)

# Axis limits
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)

# Decorazioni del grafico
plt.title(f"KNN con k={k}")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.grid()
plt.show()
```
